## 由mean shif得到候选调度点

算法：mean shift，直接调用scikit learn库中函数。关键参数为estimate_bandwidth中的quantile_value，quantile_value小，则调度点数量多，quantile_value大，则调度点数量少。

可在labels_name中确定调度点个数，原则为：商圈最远距离大和商圈包含gps点多，则调度点多，否则，少。

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

scheduling_point = {}

def labels_num(pt_num, dst):
    if dst > 7000:
        return 3
    elif dst > 6000 and pt_num > 8000:
        return 2
    else:
        return 1

process_bar = ShowProcess(len(circle_point), 'completed!')

for circle in circle_point:
    if dist[circle] > 6500:
        quantile_value = 0.1
    elif dist[circle] > 6000:
        quantile_value = 0.2
    else:
        quantile_value = 0.5
        
    n = 0
    est_n = labels_num(num[circle], dist[circle])
    
    k = 1
    m = 1
    
    if len(circle_point[circle]) < 10:
        cluster_centers = np.mean(np.array(circle_point[circle]), axis=0, keepdims=True).tolist()
    else:
    
        while n != est_n:
    
            bandwidth = estimate_bandwidth(circle_point[circle], quantile=quantile_value, n_samples=1000)
        
            if bandwidth > 0:
    
                ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
                ms.fit(circle_point[circle])
                labels = ms.labels_
                cluster_centers = ms.cluster_centers_
        
                n = len(np.unique(labels))
        
                if k % 100 == 0:
                    m *= 10
        
                if n > est_n:
                    quantile_value += 0.01/m
                elif n < est_n:
                    quantile_value -= 0.01/m
        
                k += 1
            
                if k == 500:
                    break
            
            else:
            
                n = eat_n =1
                cluster_centers = np.mean(np.array(circle_point[circle]), axis=0, keepdims=True).tolist()
        
    process_bar.show_process()      # 显示当前进度
    sleep(0.01)    
            
    scheduling_point[circle] = cluster_centers
    
    
```

## 候选调度点周围出现次数最多历史订单gps点及地址作为调度点及地址，生成txt文件交由张政写库

```python
from math import asin, cos, radians, sin, sqrt
def haversine(x, y):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)
    
    Args:
    x: [lon1, lat1]
    y: [lon2, lat2]
    """

    lon1, lat1, lon2, lat2 = x[0], x[1], y[0], y[1]
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * asin(sqrt(a))
    r = 6371
    return c * r * 1000

import io
from collections import Counter, defaultdict

def most_num_point(sch_pt, cir_pt, eps=100):
    candidate_point = []
    while True:
        for pt in cir_pt:
            d = haversine(sch_pt, pt)
            if d < eps:
                candidate_point.append(pt)
        if len(candidate_point) == 0:
            eps += 100
        else:
            break
    c = Counter([tuple(i) for i in candidate_point])
    return list(c.most_common()[0][0])

scheduling_point_new = defaultdict(list)
for circle in circle_point:
    for sch_dt in scheduling_point[circle]:
        scheduling_point_new[circle].append(most_num_point(sch_dt, circle_point[circle], 100))

df_address = df_gps
address_dict = defaultdict(dict)

df_gps_tuple = [tuple([float(gps_str.split(',')[0]), float(gps_str.split(',')[1])]) for gps_str in df_address['start_gps']]
df_address_ = list(df_address['start_address'])


for i in range(len(df_gps_tuple)):
    address_dict[df_gps_tuple[i]][str(df_address_[i])]\
    = address_dict[df_gps_tuple[i]].get(str(df_address_[i]), 0) + 1

gps_address = {}    
for gps in address_dict:
    gps_address[gps] = sorted(address_dict[gps].items(), key=lambda item: item[1], reverse=True)[0][0]

location_dict = defaultdict(dict)
df_location = list(df_address['start_location'])

for i in range(len(df_gps_tuple)):
    if df_location[i] is not np.nan:
        location_dict[df_gps_tuple[i]][str(df_location[i]).strip().split(':')[1]]\
        = location_dict[df_gps_tuple[i]].get(str(df_location[i]).strip().split(':')[1], 0) + 1
        
gps_location = {}
for gps in location_dict:
    gps_location[gps] = sorted(location_dict[gps].items(), key=lambda item: item[1], reverse=True)[0][0]
    
with io.open('/opt/test/liguangkai/cluster/scheduling/scheduling_point_changchun_location.txt', 'w', encoding="utf-8") as f:
    for circle in scheduling_point_new:
        for pt in scheduling_point_new[circle]:
            if tuple(pt) in gps_location:
                f.write('%s\t%s\t%s\n' % (circle, ','.join([str(p).decode('utf-8') for p in pt]), gps_location[tuple(pt)].decode('utf-8')))
            else:
                f.write('%s\t%s\t%s\n' % (circle, ','.join([str(p).decode('utf-8') for p in pt]), gps_address[tuple(pt)].decode('utf-8')))
```
